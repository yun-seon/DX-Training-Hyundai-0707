# (2) 연속형 확률분포

```python
# 필요한 package 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import seaborn as sns

%matplotlib inline
```

### (2-1) 정규분포 (Normal Distribution, Gaussian Distribution)

image.png

```python
# 임의의 x값 정의

x = np.linspace(-5, 5, 101)
x
```

```python
# 정규분포 수식에 따라, 확률분포 정의
# 평균 0, 표준편차 1인 정규분포 정의

y = (1 / np.sqrt(2 * np.pi)) * np.exp(- x ** 2 / 2 )
y
```

```python
# 확률 밀도 함수(pdf; probability density function) 그래프 그리기

plt.figure(figsize=(10, 6))          
plt.plot(x, y)                       
plt.xlabel("x")                      
plt.ylabel("y")                      
plt.grid()                           
plt.title("Normal Distribution without scipy")     
plt.legend(["N(0, 1)"])              
plt.show()                           
```

```python
# 확률 밀도 함수(pdf; probability density function) 그래프 그리기
# 분산 변화에 따른, 다양한 확률 밀도 함수 그래프 비교 가능

legend = []

def norm (mu, var) :
    legend.append("N(" + str(mu) + ", " + str(var) + ")")
    return sp.stats.norm(mu, var).pdf(x)        

plt.figure(figsize=(10, 6))          

for i in np.arange(4) + 1:
    plt.plot(x, norm(0, i))                
    
plt.xlabel("x")                      
plt.ylabel("y")                      
plt.grid()                           
plt.title("Normal Distribution with scipy.stats")     
plt.legend(legend)                   
plt.show()                           
```

> pdf와 cdf 차이

image.png

```python
# norm 함수를 통해 정규분포 확률 밀도 정의
# 평균 100, 표준편차 50인 정규분포 정의

rv = sp.stats.norm(loc = 100, scale = 50)
```

```python
# 특정 점에서의 확률밀도함수 값 탐색: pdf method
# x = 135에서의 확률밀도함수 값 탐색

rv.pdf(135)
```

```python
# 확률밀도함수 및 x=135인 점에서의 값 표시를 위한 그래프 시각화

x = np.linspace(0, 200, 1000)
density = rv.pdf(x)
plt.plot(x, density)

plt.scatter(135, rv.pdf(135), c = 'r') # x = 135인점 표시
plt.plot([-10, 135], [rv.pdf(135), rv.pdf(135)], 'r--')

plt.xlabel('x')
plt.ylabel('density')
plt.xlim(0, 200)

plt.show()
```

```python
# 특정 점까지의 누적확률밀도 값 탐색: cdf method
# x = 160까지의 누적분포함수 값 탐색

rv.cdf(160)
```

```python
# 누적확률분포 및 x=160인 점에서의 값 표시를 위한 그래프 시각화

x = np.linspace(0, 200, 1000)
density = rv.cdf(x)
plt.plot(x, density)

plt.scatter(160, rv.cdf(160), c = 'r') # x = 160인점 표시
plt.plot([-10, 160], [rv.cdf(160), rv.cdf(160)], 'r--')

plt.xlabel('x')
plt.ylabel('cumulative probability')
plt.xlim(0, 200)

plt.show()
```

```python
# 특정 확률까지의 누적 백분위수 탐색: ppf method
# 누적 분포 확률 값이 0.95인 x값 탐색

rv.ppf(0.95)
```

```python
# 누적확률분포의 역함수 그래프와 누적확률이 0.95인 값 표시를 위한 그래프 시각화

p = np.linspace(0, 1.0, 1000)
density = rv.ppf(p)
plt.plot(p, density)

plt.scatter(0.95, rv.ppf(0.95), c = 'r') # p = 0.95인점 표시
plt.plot([-0.05, 0.95], [rv.ppf(0.95), rv.ppf(0.95)], 'r--')

plt.xlabel('cumulative probability')
plt.ylabel('x')
plt.xlim(0, 1)

plt.show()
```

### (2-2) 지수 분포 (Exponential Distribution)

image.png

```python
# 임의의 x값 정의

x = np.linspace(0, 10, 100)
x
```

```python
# 지수분포 수식에 따라, 확률분포 정의
# lambda 1.5인 지수분포 정의

l = 1.5
y = l * np.exp(-l * x)
print(y)
```

```python
'''
# lambda 인수로 분포의 모수를 설정하고, 지수 분포를 정의
# scale: beta (1/lambda), size: 개수

x2 = np.random.exponential(scale = 0.67, loc = 0, size = 100)  # lambda 1.5를 설정하면, scale은 0.67 (=1/1.5)이 설정됨
x2
'''
```

```python
# 확률 밀도 함수(pdf; probability density function) 그래프 그리기

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title(r'$Exponential Distribution(\lambda = 1.5)$')
plt.grid()
plt.show()
```

```python
# 확률 밀도 함수(pdf; probability density function) 그래프 그리기
# lambda 값 변화에 따른, 다양한 확률 밀도 함수 그래프 비교 가능

loc = 0
lambdas = [.5, 1, 2, 4, 8]

plt.figure(figsize=(12, 8))
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Exponential Distribution')
plt.grid()

for l in lambdas:
    plt.plot(x, sp.stats.expon(loc, l).pdf(x), label=r'$(\lambda=' + str(l) + ')$')
    
plt.legend()
plt.show()
```

```python
# 누적확률 밀도 함수(cdf; cumulative density function) 그래프 그리기
# lambda 값 변화에 따른, 다양한 누적확률 밀도 함수 그래프 비교 가능

loc = 0
lambdas = [.5, 1, 2, 4, 8]

plt.figure(figsize=(12, 8))
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Exponential Distribution')
plt.grid()

for l in lambdas:
    plt.plot(x, 1 - np.exp(-l * x), label=r'$(\lambda=' + str(l) + ')$')
    
plt.legend()
plt.show()
```

```python
# lambda 값이 1.5인 지수분포의 분산

l = 1.5
print((1/l)**2) # 기댓값
```
